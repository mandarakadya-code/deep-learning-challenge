```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df.columns
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
application_df = application_df.drop(["EIN","NAME"], axis=1)
application_df.head()
```

```python
application_df["INCOME_AMT"].unique()
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
app_type_df = application_df["APPLICATION_TYPE"].value_counts().rename_axis('application_type').reset_index(name='counts')
app_type_df
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
# Loading the list with application_type values that have less than 527 value counts
application_types_to_replace = []
for index, row in app_type_df.iterrows():
    if(row['counts'] < 527):
        application_types_to_replace.append(row['application_type'])

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
counts = application_df["CLASSIFICATION"].value_counts()
counts[counts > 1]
```

```python
app_type_df = application_df["CLASSIFICATION"].value_counts().rename_axis('classification').reset_index(name='counts')
app_type_df
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
# Loading the list with classification values that have less than 1883 value counts
classifications_to_replace = []
for index, row in app_type_df.iterrows():
    if(row['counts'] < 1883):
        classifications_to_replace.append(row['classification'])
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.head()
```

```python
len(application_df.columns)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
application_dummies = pd.get_dummies(application_df[["APPLICATION_TYPE","AFFILIATION", "CLASSIFICATION", "SPECIAL_CONSIDERATIONS"]], dtype=int)
application_dummies.head()
```

```python
application_df_copy = application_df.copy()
application_df_copy = pd.concat([application_df_copy, application_dummies], axis=1)
application_df_copy = application_df_copy.drop(columns=["APPLICATION_TYPE","AFFILIATION", "CLASSIFICATION", "SPECIAL_CONSIDERATIONS","INCOME_AMT","ORGANIZATION","USE_CASE"])
application_df_copy.head()
```

```python
len(application_df_copy.columns)
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = application_df_copy["IS_SUCCESSFUL"].values
X = application_df_copy.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Attempt 3

app_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
app_cat
```

```python
application_df[app_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_df[app_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(app_cat)
encode_df.head()
```

```python
app_df_copy = application_df.copy()
app_df_copy = app_df_copy.merge(encode_df,left_index=True, right_index=True)
app_df_copy = app_df_copy.drop(app_cat,axis=1)

app_df_copy.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = app_df_copy["IS_SUCCESSFUL"].values
X = app_df_copy.drop(["IS_SUCCESSFUL"], axis=1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  80
hidden_nodes_layer2 = 30
nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)
# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))
# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
